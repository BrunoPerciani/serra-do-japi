```python
# Import Matplotlib for time series visualization with dual y-axes.
import matplotlib.pyplot as plt

# Define the analysis years.
anos = list(range(2013, 2024))

# Define average NDVI values for the municipality of Jundiaí.
ndvi_1 = [0.6368, 0.625, 0.6501, 0.6395, 0.6543, 0.6601, 0.6724, 0.6159, 0.6165, 0.638, 0.644]

# Define annual total precipitation values for the municipality of Jundiaí.
precipitacao_1 = [1342.67, 1181.88, 1655.67, 1414.02, 1397.91, 1132.26, 1376.17, 1097.22, 1039.23, 1231.17, 1273.30]

# Define average NDVI values for the Serra do Japi region.
ndvi_2 = [0.8344, 0.8158, 0.8181, 0.8322, 0.8349, 0.8116, 0.8367, 0.8093, 0.8071, 0.824, 0.842]

# Define annual total precipitation values for the Serra do Japi region.
precipitacao_2 = [1330.50, 1235.09, 1760.13, 1433.64, 1441.03, 1183.17, 1452.93, 1101.65, 1067.83, 1228.62, 1395.83]

# Create the main figure and the primary axis for NDVI visualization.
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot NDVI time series for Jundiaí and Serra do Japi on the primary y-axis.
ax1.plot(anos, ndvi_1, marker='o', linestyle='-', label='Average NDVI in the municipality of Jundiaí', color='green')
ax1.plot(anos, ndvi_2, marker='^', linestyle='--', label='Average NDVI in the Serra do Japi', color='darkgreen')

# Configure the x-axis and primary y-axis labels for NDVI.
ax1.set_xlabel('Analysis years')
ax1.set_ylabel('Average NDVI index', color='black')
ax1.tick_params('y', colors='black')

# Create a secondary y-axis to display precipitation data.
ax2 = ax1.twinx()

# Plot precipitation time series for Jundiaí and Serra do Japi on the secondary y-axis.
ax2.plot(anos, precipitacao_1, marker='s', linestyle='-.', label='Annual total precipitation in the municipality of Jundiaí', color='blue')
ax2.plot(anos, precipitacao_2, marker='d', linestyle=':', label='Annual total precipitation in the Serra do Japi', color='darkblue')

# Configure the secondary y-axis label for precipitation.
ax2.set_ylabel('Annual total precipitation (mm)', color='black')
ax2.tick_params('y', colors='black')

# Enable a standard grid to improve visual readability.
plt.grid(True, linestyle='--', linewidth=0.5, color='gray')

# Combine legends from both axes and position them below the plot.
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='upper center', bbox_to_anchor=(0.5, -0.15), shadow=True, ncol=2)

# Adjust layout spacing to prevent overlap of plot elements.
fig.tight_layout()

# Set the overall title describing NDVI and precipitation trends.
plt.title('Average NDVI and annual precipitation values in the municipality of Jundiaí and Serra do Japi (2013–2023)')

# Render the final plot.
plt.show()
```
